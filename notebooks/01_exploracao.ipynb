# JEPA-Spillover — Exploração inicial

Exploração do dataset, embeddings e espaço latente.

**Pré-requisito:** rodar o pipeline (`make pipeline` ou `python -m jepa_spillover.cli all`).

```python
import sys
from pathlib import Path

import numpy as np
import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))

from jepa_spillover.config import Config

cfg = Config.load(ROOT / 'config' / 'config.yaml')
proc = cfg.resolve('data_processed')
df = pd.read_parquet(proc / 'dataset.parquet')
print(df.shape)
df.head()
```

```python
# Distribuição por família e rótulo de spillover
df.groupby('family')['spillover_label'].agg(['count', 'mean'])
```

```python
# Espaço latente 2D (gerado por `make evaluate`)
latent = pd.read_parquet(proc / 'latent_2d.parquet')
ax = latent.plot.scatter('dim1', 'dim2', c='spillover_label', cmap='coolwarm', s=12, figsize=(8, 6))
ax.set_title('Espaço latente JEPA — risco de spillover')
```
